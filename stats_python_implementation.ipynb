```python
# compute population (actual) and sample variance for existing list `data`
arr = np.array(data)
mean = arr.mean()
population_variance = ((arr - mean) ** 2).mean()
sample_variance = ((arr - mean) ** 2).sum() / (arr.size - 1)
population_variance, sample_variance
```
